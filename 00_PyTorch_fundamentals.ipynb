### **One of the most common errors in deep learning (shape errors)**

The are two main rules that performining [matrix multiplication](https://http://matrixmultiplication.xyz/) needs to  satisfy:

The **inner dimensions** must match:
- `(3, 2) @ (3, 2)` won't work
- `(2, 3) @ (3, 2)` will work
- `(3, 2) @ (2, 3)` will work

The resulting matrix has the shape of the **outer dimensions**:
- `(2, 3) @ (3, 2)` -> `(2, 2)`
- `(3, 2) @ (2, 3)` -> `(3, 3)`

```python
torch.rand(3, 2).shape
```

```python
# don't satify rule no 1
torch.matmul(torch.rand(3, 2), torch.rand(3, 2))
```

```python
torch.matmul(torch.rand(2, 3), torch.rand(3, 2))
```

```python
# outer dimensions
torch.matmul(torch.rand(3, 2), torch.rand(2, 3))
```

[ matrix multiplication](http//matrixmultiplication.xyz/)

```python
# Shape for matrix multiplication

tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])

# torch.mm(tensor_A, tensor_B) # torch.mm is the same as torch.matmul (it's an alias for writing less code)
torch.matmul(tensor_A, tensor_B)
```

```python
tensor_A.shape, tensor_B.shape
```


To fix our tensor shape issues, we can manipulate the shape  of ones of our tensors using a **Transpose**.     
A **Transpose** switches the axes or dimensions of a given tensor.

```python
tensor_B, tensor_B.shape
```

```python
tensor_B.T, tensor_B.T.shape
```

image.png

```python
# The matrix multiplication operation works when tensor_B is transposed

print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}")
#print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

### **Reshaping, stacking, squeezing and unsqueezing tensors**

- Reshaping - reshapes an input tensor to a defined shape.
- View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor.
- Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack). `torch stack, torch vstack, torch hstack`.
- Squeeze - removes all `1` dimensions from a tensor
- Unsqueeze - adds a `1` dimensions to a target tensor
- Permute - Return a view of the input with dimensions permuted (swaped) in a certain way.

```python
# Create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

Now let's add an extra dimension with torch.reshape().

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 7)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9, 1)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the original input)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0) # try changing dim to dim=1 and see what happens
x_stacked
```

```python
# try changing dim to dim=1 and see what happens
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

# `torch.permute` - rearranges the dimensions of a target tensor in a specified order

```python
# Create tensor with specific shape
x_original = torch.rand(size=(224, 224, 3)) # [height, width, colour_channels]

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}") # [colour_channels, height, width]
```

```python
x_original[0, 0, 0] = 728218
x_original[0, 0, 0], x_permuted[0, 0, 0]
```

## Reproducibility (trying to take the random out of random)

In short how neural network learns:

`start with random numbers -> tensor operations -> update random numbers to try and make them better representation of the data -> (again and again and again)`

To reduce randomness in neural networks and PyTorch comes the concept **random seed**.

Essentially what the random seed does is "flavour" the randomness.

```python
import torch

# Create two random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
random_tensor_A == random_tensor_B
```

```python
1 == 1
```

## Let's make some random but reproducible tensors

```python
import torch

# # Set the random seed

RANDOM_SEED=42 # try changing this to different values and see what happens to the numbers below

torch.manual_seed(seed=RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(seed=RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)

random_tensor_C == random_tensor_D
```
